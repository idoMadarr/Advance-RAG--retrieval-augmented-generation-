```python
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from IPython.display import Markdown, display
import base64
```

```python
vision_model = ChatOllama(model="qwen2.5vl:7b")
llm = ChatOllama(model="gemma3:12b")
```

```python
# Load image
with open("release_notes.png", "rb") as f:
    image_base64 = base64.b64encode(f.read()).decode("utf-8")
```

```python
raw_response = vision_model.invoke([
    HumanMessage(content=[
        {"type": "text", "text": "Extract all text from this image. Return only the extracted text."},
        {"type": "image_url", "image_url": f"data:image/jpeg;base64,{image_base64}"}
    ])
])

print(raw_response.content)
```

```python
release_notes_prompt = ChatPromptTemplate.from_template("""
    You are an experienced Product Manager responsible for writing professional release notes.
    Your task is to convert the provided development tasks into clear, concise, customer-facing release notes.

    Instructions:
    - Write in a professional and positive tone.
    - Focus on user impact rather than implementation details.
    - Group related changes together when appropriate.
    - Categorize items under:
      - New Features
      - Improvements
      - Bug Fixes
    - Rewrite technical task descriptions into language that non-technical users can understand.
    - Do not invent features or functionality that are not mentioned.
    - Use bullet points.
    - Response only in Hebrew

    Tasks:
    {release_notes}

    Return only the release notes in Markdown format.
""")
```

```python
chain = release_notes_prompt | llm |  StrOutputParser()
```

```python
response = chain.invoke({
    "release_notes": raw_response.content
})

```

```python
display(Markdown(response))
```
