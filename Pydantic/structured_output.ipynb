# Structured Outputs
#### Structured Outputs is a feature that ensures the model will always generate responses that adhere to your supplied JSON Schema, so you don’t need to worry about the model omitting a required key, or hallucinating an invalid enum value.

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from pydantic import BaseModel, Field
from langchain_ollama import ChatOllama

load_dotenv()
```

```python
# Create LLM:
# Notice that not all the models support jsonschema support

model = init_chat_model(model="openai:gpt-4o-mini")
# llm = ChatOllama(model="gemma3:12b")
```

```python
class Actor(BaseModel):
    name: str
    role: str

class Movie(BaseModel):
    title: str = Field("The title of the movie")
    description: str = Field("Short description of the movie")
    year: int = Field("Release year")
    director: str = Field("Director of the movie")
    rating: float = Field("Rating of the movie")
    cast: list[Actor] = []
```

```python
pydantic_llm = model.with_structured_output(Movie)
```

```python
pydantic_llm.invoke("Proving me some details about the movie Load Of The Rings")
```
