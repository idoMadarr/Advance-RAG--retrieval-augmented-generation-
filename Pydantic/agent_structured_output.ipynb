# Structured Outputs for Agent

```python
import os
from dotenv import load_dotenv
from langchain.agents import create_agent
from pydantic import BaseModel, Field

load_dotenv()
```

```python
class ContactInfo(BaseModel):
    """ Contact unformation for a person """
    name: str = Field("The name of the contact")
    email: str = Field("The email of the contact")
    phone: str = Field("The phone number of the contact")
```

```python
agent = create_agent(
    model="openai:gpt-4o-mini",
    response_format=ContactInfo
)
```

```python
result = agent.invoke({
    "messages": [{ "role": "user", "content": "extract contact info from: Jone Doe, hone@gmail.com, 052-5544747" }]
})

result["structured_response"]
```
